````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Time configuration (6 years of monthly data)
dates = pd.date_range(start='2018-01-01', end='2023-12-01', freq='MS')
n = len(dates)

# 2. Components creation
# Trend: Linear growth of sales
trend = np.linspace(2000, 5000, n)

# Seasonality: Sine wave (peak in summer, trough in winter)
# Amplitude of 1500 units
seasonality = 1500 * np.sin(2 * np.pi * dates.month / 12 - (np.pi / 2))

# Noise: Random fluctuations (weather, minor disruptions, etc.)
noise = np.random.normal(0, 200, n)

# 3. Assembling the data
# Adding 2000 baseline to ensure no negative sales occur
sales = trend + seasonality + noise + 2000
sales = sales.astype(int)

# 4. Creating DataFrame
df = pd.DataFrame({
'Date': dates,
'Sales': sales
})

# 5. Saving to CSV and preview
filename = 'ice_cream_sales.csv'
df.to_csv(filename, index=False)
print(f"File '{filename}' has been created successfully!")
display(df.head())

# Quick visualization
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Sales'], label='Ice Cream Sales')
plt.title('Generated Historical Data: Ice Cream Sales (2018-2023)')
plt.xlabel('Date')
plt.ylabel('Sales Volume')
plt.legend()
plt.grid(True)
plt.show()
````

## Split Data

### Subtask:
Split the `df` DataFrame into training and test sets. The training set will include data up to the end of 2022, and the test set will include data from 2023.


**Reasoning**:
The 'Date' column needs to be converted to datetime objects to allow for proper filtering by date. This ensures that the split into training and test sets is accurate.



````python
df['Date'] = pd.to_datetime(df['Date'])
print("Date column converted to datetime.")
````

**Reasoning**:
Splitting the `df` DataFrame into training and test sets using date-based filtering, as instructed.



````python
train_df = df[df['Date'] <= '2022-12-31']
test_df = df[df['Date'] >= '2023-01-01']

print("DataFrame split into training and test sets.")
print(f"Training set shape: {train_df.shape}")
print(f"Test set shape: {test_df.shape}")
````

**Reasoning**:
The next step is to convert the 'Date' column in both the training and test sets into a numerical format suitable for Linear Regression, as per the subtask instructions. Calculating the number of months since the first date of the training set is a common and effective way to represent time numerically.



````python
first_date = train_df['Date'].min()
train_df['Month_Elapsed'] = (train_df['Date'].dt.year - first_date.year) * 12 + (train_df['Date'].dt.month - first_date.month)
test_df['Month_Elapsed'] = (test_df['Date'].dt.year - first_date.year) * 12 + (test_df['Date'].dt.month - first_date.month)

print("Date columns converted to numerical 'Month_Elapsed' feature.")
print("First 5 rows of train_df with 'Month_Elapsed':")
display(train_df.head())
print("First 5 rows of test_df with 'Month_Elapsed':")
display(test_df.head())
````

**Reasoning**:
The previous code generated a `SettingWithCopyWarning` because `train_df` and `test_df` were views of the original DataFrame. To prevent this, I will explicitly create copies of the filtered DataFrames using `.copy()` before adding the 'Month_Elapsed' column.



````python
first_date = train_df['Date'].min()
train_df = train_df.copy()
test_df = test_df.copy()

train_df['Month_Elapsed'] = (train_df['Date'].dt.year - first_date.year) * 12 + (train_df['Date'].dt.month - first_date.month)
test_df['Month_Elapsed'] = (test_df['Date'].dt.year - first_date.year) * 12 + (test_df['Date'].dt.month - first_date.month)

print("Date columns converted to numerical 'Month_Elapsed' feature.")
print("First 5 rows of train_df with 'Month_Elapsed':")
display(train_df.head())
print("First 5 rows of test_df with 'Month_Elapsed':")
display(test_df.head())
````

**Reasoning**:
Now that the training and test sets are prepared with a numerical 'Month_Elapsed' feature, the next step is to train a Linear Regression model. The 'Month_Elapsed' column will serve as the independent variable, and 'Sales' will be the dependent variable.



````python
from sklearn.linear_model import LinearRegression

# Prepare the data for Linear Regression
X_train = train_df[['Month_Elapsed']]
y_train = train_df['Sales']

# Initialize and train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

print("Linear Regression model trained successfully.")
print(f"Model intercept: {model.intercept_:.2f}")
print(f"Model coefficient (slope): {model.coef_[0]:.2f}")
````

**Reasoning**:
The trained Linear Regression model needs to be used to forecast sales for the test set. I will use the 'Month_Elapsed' column from the `test_df` as the input for the `predict` method of the trained model.



````python
X_test = test_df[['Month_Elapsed']]
predictions = model.predict(X_test)

test_df['Forecasted_Sales'] = predictions

print("Sales forecasted for the test set.")
print("First 5 rows of test_df with 'Forecasted_Sales':")
display(test_df.head())
````

**Reasoning**:
To visualize the results, I need to plot the original training data, the actual test data, and the forecasted sales from the Linear Regression model. This will clearly show how the model performs against historical and future data.



````python
plt.figure(figsize=(14, 7))
plt.plot(train_df['Date'], train_df['Sales'], color='black', label='Original Training Data')
plt.plot(test_df['Date'], test_df['Sales'], color='green', label='Actual Test Data (2023)')
plt.plot(test_df['Date'], test_df['Forecasted_Sales'], color='red', linestyle='--', label='Linear Regression Forecast (2023)')

plt.title('Ice Cream Sales: Linear Regression Forecast vs. Actuals')
plt.xlabel('Date')
plt.ylabel('Sales Volume')
plt.legend()
plt.grid(True)
plt.show()

print("Visualization of training data, actual test data, and linear regression forecast displayed.")
````

### Linear Regression Model Performance Summary

The Linear Regression model, trained on the 'Month_Elapsed' feature, provides a simple linear trend for forecasting ice cream sales.

**Observations:**
- **Trend Capture:** The model successfully captures the overall increasing trend in sales over time, as evidenced by its positive slope.
- **Seasonality:** However, the straight-line forecast **does not capture the inherent seasonality** of ice cream sales (peak in summer, trough in winter). As seen in the visualization, the actual sales data fluctuates significantly above and below the linear forecast in both the training and test sets, especially during peak and off-peak seasons.
- **Accuracy:** During 2023, the forecast consistently underpredicts sales during the summer months and overpredicts during the winter months, highlighting its inability to account for seasonal patterns.

In conclusion, while Linear Regression is effective for capturing general trends, it is not suitable for data with strong seasonal components like ice cream sales. More sophisticated time series models (e.g., ARIMA, Prophet, or models incorporating seasonal dummy variables) would be required to accurately forecast such patterns.

## Train Holt-Winters Model

### Subtask:
Train a Holt-Winters Exponential Smoothing model on the `train_df` DataFrame, configured to account for both trend and seasonality.


**Reasoning**:
To train a Holt-Winters Exponential Smoothing model, I need to import the necessary class, instantiate it with the specified parameters for trend and seasonality, then fit it to the training data and display a summary.



````python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Instantiate the Holt-Winters model
# Assuming an additive trend and additive seasonality based on the nature of ice cream sales
# Seasonal period is 12 for monthly data
model_hw = ExponentialSmoothing(
    train_df['Sales'],
    seasonal_periods=12,
    trend='add',
    seasonal='add'
)

# Fit the model to the training data
fit_hw = model_hw.fit()

print("Holt-Winters Exponential Smoothing model trained successfully.")
print("Model Summary:")
print(fit_hw.summary())
````

## Make Forecast

**Reasoning**:
I need to use the `forecast()` method of the `fit_hw` object to generate predictions for the test set period and then store these predictions in a new column in `test_df`.



````python
forecast_hw = fit_hw.forecast(steps=len(test_df))
test_df['Forecasted_Sales_HW'] = forecast_hw

print("Holt-Winters sales forecasts generated for the test set.")
print("First 5 rows of test_df with 'Forecasted_Sales_HW':")
display(test_df.head())
````

**Reasoning**:
To complete the subtask, I need to visualize the original training data, actual test data, and the Holt-Winters forecast on a single plot to compare its performance against the actuals and the previous linear regression model.



````python
plt.figure(figsize=(14, 7))
plt.plot(train_df['Date'], train_df['Sales'], color='black', label='Original Training Data')
plt.plot(test_df['Date'], test_df['Sales'], color='green', label='Actual Test Data (2023)')
plt.plot(test_df['Date'], test_df['Forecasted_Sales_HW'], color='blue', linestyle='--', label='Holt-Winters Forecast (2023)')

plt.title('Ice Cream Sales: Holt-Winters Forecast vs. Actuals')
plt.xlabel('Date')
plt.ylabel('Sales Volume')
plt.legend()
plt.grid(True)
plt.show()

print("Visualization of training data, actual test data, and Holt-Winters forecast displayed.")
````

````python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# Apply seasonal differencing (D=1, S=12)
train_df['Sales_Seasonal_Diff'] = train_df['Sales'].diff(12)

# Plot ACF and PACF of the seasonally differenced series
fig, axes = plt.subplots(2, 1, figsize=(12, 10))

plot_acf(train_df['Sales_Seasonal_Diff'].dropna(), ax=axes[0], lags=29, title='ACF of Seasonally Differenced Sales Data')
plot_pacf(train_df['Sales_Seasonal_Diff'].dropna(), ax=axes[1], lags=29, title='PACF of Seasonally Differenced Sales Data')

plt.tight_layout()
plt.show()

print("ACF and PACF plots of the seasonally differenced sales data displayed.")
````

````python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# Apply seasonal differencing (D=1, S=12)
train_df['Sales_Seasonal_Diff'] = train_df['Sales'].diff(12)

# Plot ACF and PACF of the seasonally differenced series
fig, axes = plt.subplots(2, 1, figsize=(12, 10))

# Adjust lags based on the new sample size (60 - 12 = 48 observations -> max lag is 23)
plot_acf(train_df['Sales_Seasonal_Diff'].dropna(), ax=axes[0], lags=23, title='ACF of Seasonally Differenced Sales Data')
plot_pacf(train_df['Sales_Seasonal_Diff'].dropna(), ax=axes[1], lags=23, title='PACF of Seasonally Differenced Sales Data')

plt.tight_layout()
plt.show()

print("ACF and PACF plots of the seasonally differenced sales data displayed with adjusted lags.")
````

*   **Non-Seasonal AR(p) and MA(q) Components:**
    *   **ACF of Seasonally Differenced Series:** There is a significant spike at lag 1, followed by a relatively quick drop-off. This suggests a non-seasonal Moving Average (MA) component of order 1 (`q=1`).

## Train SARIMA Model

### Subtask:
Train a Seasonal AutoRegressive Integrated Moving Average (SARIMA) model on the `train_df` DataFrame using the determined orders. The model should explicitly account for both trend and seasonality.


**Reasoning**:
Now that the ACF and PACF plots have been analyzed (implicitly in the instructions by providing specific orders), I will proceed to train the SARIMA model with the given non-seasonal (p, d, q) and seasonal (P, D, Q, S) orders. This involves importing `SARIMAX`, instantiating the model, fitting it to the training data, and then printing the summary.



````python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Instantiate the SARIMA model
# Non-seasonal order (p, d, q) = (1, 0, 1)
# Seasonal order (P, D, Q, S) = (1, 1, 1, 12)
model_sarima = SARIMAX(train_df['Sales'],
                      order=(1, 0, 1),
                      seasonal_order=(1, 1, 1, 12),
                      enforce_stationarity=False,
                      enforce_invertibility=False)

# Fit the SARIMA model to the training data
fit_sarima = model_sarima.fit(disp=False)

print("SARIMA model trained successfully.")
print("Model Summary:")
print(fit_sarima.summary())
````

The previous SARIMA model fit resulted in a 'ConvergenceWarning' and a singular covariance matrix, leading to unstable standard errors (NaNs) and an astronomically large coefficient. This indicates a numerical instability or over-parameterization, likely stemming from the chosen orders (1,0,1) for non-seasonal components. A common and more stable SARIMA configuration for data with both trend and seasonality, which includes both non-seasonal and seasonal differencing (d=1, D=1), is (0, 1, 1) for non-seasonal components and (0, 1, 1, 12) for seasonal components. This simplified model is less likely to suffer from over-parameterization and often converges more reliably.



````python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Instantiate the SARIMA model with refined orders to improve convergence and stability
# Non-seasonal order (p, d, q) = (0, 1, 1) - applies a first difference (d=1) to handle trend
# Seasonal order (P, D, Q, S) = (0, 1, 1, 12) - applies seasonal difference (D=1) for seasonality
model_sarima = SARIMAX(train_df['Sales'],
                      order=(0, 1, 1),
                      seasonal_order=(0, 1, 1, 12),
                      enforce_stationarity=False, # Keep for now, but consider removing if issues persist
                      enforce_invertibility=False) # Keep for now, but consider removing if issues persist

# Fit the SARIMA model to the training data
fit_sarima = model_sarima.fit(disp=False)

print("SARIMA model trained successfully with revised orders.")
print("Model Summary:")
print(fit_sarima.summary())
````

**Reasoning**:
The previous SARIMA model fit continued to exhibit a 'ConvergenceWarning' and singular covariance matrix even with revised orders. This often happens when `enforce_stationarity` and `enforce_invertibility` are set to `False`, allowing the model to explore non-stationary/non-invertible parameter space which can lead to instability during optimization. By removing these parameters, the model will default to `True`, which can help ensure more stable and interpretable parameter estimates and improve convergence.



````python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Instantiate the SARIMA model with revised orders to improve convergence and stability
# Non-seasonal order (p, d, q) = (0, 1, 1) - applies a first difference (d=1) to handle trend
# Seasonal order (P, D, Q, S) = (0, 1, 1, 12) - applies seasonal difference (D=1) for seasonality
model_sarima = SARIMAX(train_df['Sales'],
                      order=(0, 1, 1),
                      seasonal_order=(0, 1, 1, 12))

# Fit the SARIMA model to the training data
fit_sarima = model_sarima.fit(disp=False)

print("SARIMA model trained successfully with revised orders and default stationarity/invertibility enforcement.")
print("Model Summary:")
print(fit_sarima.summary())
````

**Reasoning**:
The previous SARIMA model fits repeatedly produced warnings about non-invertible seasonal moving average and singular covariance matrices, indicating instability and poorly identified parameters, particularly for the moving average terms. To resolve these issues and obtain a stable model, I will significantly simplify the SARIMA orders by setting all autoregressive (p, P) and moving average (q, Q) components to zero. This results in a SARIMA(0,1,0)(0,1,0,12) model, which relies solely on non-seasonal (d=1) and seasonal (D=1, S=12) differencing to account for trend and seasonality, thus addressing the previous numerical challenges.



````python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Instantiate the SARIMA model with simplified orders to improve stability
# Non-seasonal order (p, d, q) = (0, 1, 0) - relies only on first differencing for trend
# Seasonal order (P, D, Q, S) = (0, 1, 0, 12) - relies only on seasonal differencing for seasonality
model_sarima = SARIMAX(train_df['Sales'],
                      order=(0, 1, 0),
                      seasonal_order=(0, 1, 0, 12))

# Fit the SARIMA model to the training data
fit_sarima = model_sarima.fit(disp=False)

print("SARIMA model trained successfully with simplified orders.")
print("Model Summary:")
print(fit_sarima.summary())
````

## Calculate Metrics for Linear Regression

### Subtask:
Calculate Mean Absolute Error (MAE), Root Mean Squared Error (RMSE), and Mean Absolute Percentage Error (MAPE) for the Linear Regression model using the actual sales and its forecasts.


**Reasoning**:
To calculate the required evaluation metrics (MAE, RMSE, MAPE) for the Linear Regression model, I need to import the necessary functions from `sklearn.metrics` and define a custom MAPE function, then apply them to the actual and forecasted sales data.



````python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

def mean_absolute_percentage_error(y_true, y_pred):
    # Replace zeros in y_true with a small epsilon to avoid division by zero
    # Or filter out corresponding predictions if sales can truly be zero
    # For this dataset, sales are always positive, so direct calculation is fine.
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Extract actual and forecasted sales for Linear Regression
y_true_lr = test_df['Sales']
y_pred_lr = test_df['Forecasted_Sales']

# Calculate MAE
mae_lr = mean_absolute_error(y_true_lr, y_pred_lr)

# Calculate RMSE
rmse_lr = np.sqrt(mean_squared_error(y_true_lr, y_pred_lr))

# Calculate MAPE
mape_lr = mean_absolute_percentage_error(y_true_lr, y_pred_lr)

print("Linear Regression Model Metrics:")
print(f"  Mean Absolute Error (MAE): {mae_lr:.2f}")
print(f"  Root Mean Squared Error (RMSE): {rmse_lr:.2f}")
print(f"  Mean Absolute Percentage Error (MAPE): {mape_lr:.2f}%")
````
